```python
#import packages
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#read the data
df1 = pd.read_csv('new_train.csv')
```

```python
df1.head()
```

```python
df1['ApplicantIncome'].hist(alpha=0.6, bins=20, grid=False, figsize=(8,6), label='Applicant Income')
df1['CoapplicantIncome'].hist(alpha=0.6, bins=20, grid=False, label='Co-Applicant Income')

plt.legend()
plt.axis(xmin=0,xmax=40000)
```

```python
df1['Total_income'] = df1['ApplicantIncome']+df1['CoapplicantIncome']
df1['Total_income'].hist(alpha=0.6, bins=20, grid=False, label='Total Income')

plt.legend()
plt.axis(xmin=0,xmax=40000)
```

```python
df1['log10_income'] = np.log10(df1['Total_income'])
df1['log10_LoanAmount'] = np.log10(df1['LoanAmount'])
```

```python
fig , axs = plt.subplots(ncols=2, figsize=(12,4))

df1['Total_income'].hist(histtype='step', bins=20, grid=False, density=True, ax=axs[0])
df1['Total_income'].plot.density(bw_method=0.5, ax=axs[0])

df1['log10_income'].hist(histtype='step', bins=20, grid=False, density=True, ax=axs[1])
df1['log10_income'].plot.density(bw_method=0.5, ax=axs[1])


axs[0].axis(xmin=-20000,xmax=40000)
axs[0].set_title('Distribution of Total income')
axs[1].axis(xmin=2,xmax=6)
axs[1].set_title('Distribution of Total income(Log_10)')
```

```python
fig , axs = plt.subplots(ncols=2, figsize=(12,4))
df1['LoanAmount'].hist(histtype='step', bins=20, grid=False, density=True, ax=axs[0])
df1['LoanAmount'].plot.density(bw_method=0.5,ax=axs[0])

df1['log10_LoanAmount'].hist(histtype='step', bins=20, grid=False, density=True,ax=axs[1])
df1['log10_LoanAmount'].plot.density(bw_method=0.5,ax=axs[1])

axs[0].axis(xmin=-200, xmax=800)
axs[0].set_title('Distribution of Loan Amount')
axs[1].axis(xmin=1.0, xmax=3.5)
axs[1].set_title('Distribution of Loan Amount(Log_10)')
```
